## Triangulation versus quadrangulation

This section explores the utility of the `seed_point_linking` parameter. By default, growbikenet uses a triangulation between seed points on a grid. However, some cities have a very grid-like street network, like Manhattan or Barcelona, where a quadrangulation makes more sense. To achieve such a quadrangulation that covers the city with a similar density, use `seed_point_linking = "quadrangulate"` and lower the default `seed_point_grid_spacing` of 1707 to 1000:

```python
edges_ranked = gbn.growbikenet("Barcelona",
                               seed_point_grid_spacing=1000,
                               seed_point_linking = "quadrangulate")
```

Explore the results interactively:

```python
edges_ranked.explore(tiles="CartoDB Positron", style_kwds={"weight": 4, "color": "#096a51"})
```

Compare this to the output of the default triangulation which features some ugly zig-zags:

```python
edges_ranked = gbn.growbikenet("Barcelona")
edges_ranked.explore(tiles="CartoDB Positron", style_kwds={"weight": 4, "color": "#096a51"})
```
